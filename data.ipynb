```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# API request
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=IBM&interval=5min&apikey=F73EEQ0IIOVK7NSJ'
response = requests.get(url)
data = response.json()

# Extract the time series data
time_series = data.get("Time Series (5min)", {})

# Convert to a DataFrame
df = pd.DataFrame.from_dict(time_series, orient="index", dtype=float)
df = df.rename(columns={
    "1. open": "Open",
    "2. high": "High",
    "3. low": "Low",
    "4. close": "Close",
    "5. volume": "Volume"
})
df.index = pd.to_datetime(df.index)
df = df.sort_index()  # Ensure the data is sorted by time

# Plot the closing prices
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Closing Price')
plt.title('IBM Stock Price (5-Min Interval)')
plt.xlabel('Time')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```
